```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")

BackOrders_data_num_impute = imputer.fit_transform(BackOrders_data_num)
```

```python
BackOrders_data_num_impute_df = pd.DataFrame(BackOrders_data_num_impute, columns=BackOrders_data_num_col)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for col in BackOrders_data_cat.columns :
    BackOrders_data_cat[col] = encoder.fit_transform(BackOrders_data_cat[col])
```

```python
BackOrders_data_cat.head()
```

```python
target_variable
```

```python
target_variable = encoder.fit_transform(target_variable).copy()
target_variable = pd.Series(target_variable)
```

```python
target_variable
```

```python
total_data=pd.concat([BackOrders_data_num_impute_df, BackOrders_data_cat], axis=1)
```

```python
total_data.head()
```

```python
total_data.info()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(total_data, target_variable, test_size=0.25, random_state=42)
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(random_state=42)

rfe_RF=RFE(RF,n_features_to_select=3,verbose=21)

rfe_RF.fit(x_train, y_train)
names=x_train.columns.tolist()
b=rfe_RF.ranking_

a=pd.DataFrame(sorted(list(map(lambda x,y: (x,y),b,names))),columns=['rank','features'])
a.to_csv('feature_selection.csv')

RFE_selected=a['features'][a['rank']==1]
```

```python
RFE_selected
```

```python
x_train=x_train[RFE_selected]
x_test=x_test[RFE_selected]
```

```python
x_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
RF_model=RandomForestClassifier(max_depth=3, max_features=0.6, min_samples_leaf=4,n_estimators=9)
RF_model.fit(x_train,y_train)
```

```python
pred=RF_model.predict(x_test)
```

```python
out_data=pd.DataFrame({'true':y_test,'model_predicted':pred})
out_data.to_csv('BackOrders_Prediction_Analysis.csv')
```

```python
preds=RF_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_test=accuracy_score(y_test, preds)
accuracy_test
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"n_estimators" : [100, 200, 300],
              "max_depth" : [3, 6],
              "min_samples_leaf" : [2, 4],
              "max_features" : [0.4,0.5]
             }

RF_model=RandomForestClassifier()
clf = GridSearchCV(RF_model, param_grid, cv=3)
clf.fit(x_train,y_train)
clf.best_estimator_
```

```python
preds=clf.best_estimator_.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_test=accuracy_score(y_test, preds)
```
